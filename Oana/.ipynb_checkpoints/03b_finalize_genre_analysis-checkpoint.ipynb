# Phase 2: Genre Trends Cleanup & Visualization (Local)
Processing the aggregated genre counts from the HDFS run. 
filtering noise and generating final plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_theme(style="whitegrid")

# Load Aggregated Data
dfs = []
csv_files = ["genre_trends_2010_2015.csv", "genre_trends_2015_2020.csv", "genre_trends_2020_2025.csv"]

for f in csv_files:
    try:
        temp_df = pd.read_csv(f)
        dfs.append(temp_df)
        print(f"✅ Loaded {f}: {len(temp_df)} records")
    except FileNotFoundError:
        print(f"⚠️ {f} not found. Skipping.")

if dfs:
    df = pd.concat(dfs, ignore_index=True)
    # Remove duplicates if intervals overlapped (2015 might be in both)
    df = df.drop_duplicates(subset=["read_year", "genre"])
    print(f"\nMerged Data: {len(df)} records.")
    print(f"Year Range: {df['read_year'].min()} to {df['read_year'].max()}")
else:
    print("❌ No data files found. Please run the HDFS extraction first.")
    # Create dummy data for testing
    print("Creating dummy dataframe for code validation...")
    data = []
    for y in range(2010, 2026):
        data.append({"read_year": y, "genre": "fantasy", "count": (y-2009)*100})
        data.append({"read_year": y, "genre": "romance", "count": (y-2009)*120})
    df = pd.DataFrame(data)
```

```python
# Cleanup Logic
if not df.empty:
    ignore_list = ["have", "favorite-books", "books-i-own", "owned-books", "library", "default", "favorites"]
    df_clean = df[~df["genre"].isin(ignore_list)].copy()

    # Merge 'ya' -> 'young-adult'
    df_clean["genre"] = df_clean["genre"].replace({"ya": "young-adult"})

    # Re-aggregate (in case merge created duplicates)
    df_clean = df_clean.groupby(["read_year", "genre"], as_index=False)["count"].sum()
else:
    df_clean = df
```

## Interval Analysis
Analyzing genre trends across three specific time intervals:
1. 2010 - 2015
2. 2015 - 2020
3. 2020 - 2025

```python
def analyze_interval(df, start_year, end_year, ax=None):
    """Filters data for an interval, finds top genres, and plots trends."""
    # Filter inclusive of start and end year
    subset = df[(df["read_year"] >= start_year) & (df["read_year"] <= end_year)]
    
    if subset.empty:
        if ax:
            ax.text(0.5, 0.5, f"No Data for {start_year}-{end_year}", 
                    horizontalalignment='center', verticalalignment='center', transform=ax.transAxes)
            ax.set_title(f"Top Genres: {start_year}-{end_year}")
        print(f"⚠️ No data found for interval {start_year}-{end_year}")
        return
    
    # Identify Top 5 Genres for this specific interval
    top_genres_interval = subset.groupby("genre")["count"].sum().nlargest(5).index.tolist()
    subset_plot = subset[subset["genre"].isin(top_genres_interval)]
    
    # Plotting
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 5))
    
    sns.lineplot(data=subset_plot, x="read_year", y="count", hue="genre", marker="o", ax=ax, linewidth=2.5)
    ax.set_title(f"Top Genres: {start_year}-{end_year}")
    ax.set_ylabel("Books Read")
    ax.set_xlabel("Year")
    
    # Set integer ticks for years found in data
    years_in_data = sorted(subset_plot["read_year"].unique())
    if years_in_data:
        ax.set_xticks(years_in_data)
        
    ax.legend(title="Genre")

# Define 3 Intervals (Overlapping endpoints as requested? usually intervals are [start, end))
# Request: 2010-2015, 2015-2020, 2020-2025
intervals = [
    (2010, 2015),
    (2016, 2020),
    (2021, 2025)
]

# Create Subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=False)

for i, (start, end) in enumerate(intervals):
    analyze_interval(df_clean, start, end, ax=axes[i])

plt.tight_layout()
plt.savefig("genre_trends_intervals_5yr.png")
plt.show()
```
